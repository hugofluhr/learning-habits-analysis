# Filter SPM Results Tables

Takes a whole-brain uncorrected (p < 0.001) SPM results CSV and displays only the clusters that survive FWE correction at the cluster level (p < 0.05).

```python
import pandas as pd
```

```python
def filter_fwe(path, fwe_threshold=0.05):
    """Load an SPM results CSV and return only FWE-surviving cluster peaks.

    Expects the fixed single-header format from the spm_list export script.
    Columns (positional): p, c, p(FWE-corr)[cluster], p(FDR-corr)[cluster],
                          equivk, p(unc)[cluster], p(FWE-corr)[peak],
                          p(FDR-corr)[peak], T, equivZ, p(unc)[peak], x, y, z
    """
    df = pd.read_csv(path, header=0, index_col=False)

    cluster_fwe = pd.to_numeric(df.iloc[:, 2], errors='coerce')
    cluster_k   = pd.to_numeric(df.iloc[:, 4], errors='coerce')
    peak_T      = pd.to_numeric(df.iloc[:, 8], errors='coerce')
    peak_Z      = pd.to_numeric(df.iloc[:, 9], errors='coerce')
    peak_p      = pd.to_numeric(df.iloc[:, 10], errors='coerce')
    x           = pd.to_numeric(df.iloc[:, 11], errors='coerce')
    y           = pd.to_numeric(df.iloc[:, 12], errors='coerce')
    z           = pd.to_numeric(df.iloc[:, 13], errors='coerce')

    mask = cluster_fwe < fwe_threshold

    out = pd.DataFrame({
        'x':              x[mask].values,
        'y':              y[mask].values,
        'z':              z[mask].values,
        'k':              cluster_k[mask].values,
        'T':              peak_T[mask].round(2).values,
        'Z':              peak_Z[mask].round(2).values,
        'p(unc)':         peak_p[mask].apply(lambda v: f'{v:.2e}').values,
        'p(FWE) cluster': cluster_fwe[mask].apply(lambda v: f'{v:.2e}').values,
    })

    return out.reset_index(drop=True)
```

## PPI model

```python
path = '/Users/hugofluhr/phd_local/data/LearningHabits/spm_outputs_noSDC/PPI/gppi_putamen_Hvalchosen_deconv_2026-03-18-07-39-25/second-lvl/contrast-01_second_stimxhval_chosen1/whole_brain_uncorrected_001.csv'
result = filter_fwe(path)
print(f'{len(result)} cluster(s) surviving FWE p < 0.05')
display(result)
```
