```python
import os
import numpy as np
from matplotlib import pyplot as plt
from nilearn import image
import pandas as pd
```

```python
sfnr_dir = '/Users/hugofluhr/phd_local/data/LearningHabits/sfnr'
outdir = os.path.join(sfnr_dir, 'group_means')

if not os.path.exists(outdir):
    os.makedirs(outdir)
```

```python
nii_files = [os.path.join(root, f) 
             for root, _, files in os.walk(sfnr_dir) 
             for f in files if f.endswith('.nii')]
nii_files.sort()
```

```python
sub_run_files = {}
for f in nii_files:
    sub = os.path.basename(os.path.dirname(os.path.dirname(f)))
    run = os.path.basename(os.path.dirname(f))
    sub_run_files.setdefault(sub, {})[run] = f

subjects = sorted(sub_run_files.keys())
run_labels = ['run-1', 'run-2', 'run-3']

first_img = image.load_img(sub_run_files[subjects[0]][run_labels[0]])
vol_shape = first_img.shape

data = np.zeros((len(subjects), len(run_labels), *vol_shape), dtype=np.float32)

for i, sub in enumerate(subjects):
    for j, run in enumerate(run_labels):
        img = image.load_img(sub_run_files[sub][run])
        data[i, j] = img.get_fdata()
```

## Load ROI masks

```python
roi_dir = '/Users/hugofluhr/phd_local/data/LearningHabits/dev_sample/masks/MNI152NLin2009cAsym'

def load_rois(roi_dir, ref_img = None, bool_data = False):
    files = [f for f in os.listdir(roi_dir) if f.endswith('.nii')]
    roi_names = [f.split('_')[0] for f in files]

    rois = {n: image.load_img(os.path.join(roi_dir, f)) for n, f in zip(roi_names, files)}

    if ref_img is not None:
        rois = {n: image.resample_to_img(img, ref_img, force_resample=True, copy_header=True, interpolation='nearest')
            for n, img in rois.items()}

    if bool_data:
        rois = {n: np.squeeze(img.get_fdata() > 0)
            for n, img in rois.items()}

    return rois

rois = load_rois(roi_dir, ref_img=first_img, bool_data=True)
roi_names = rois.keys()
```

```python
# Compute mean SFNR in each ROI for each subject/run
mean_sfnr = np.zeros((len(subjects), len(run_labels), len(roi_names)))

for i, sub in enumerate(subjects):
    for j, run in enumerate(run_labels):
        sfnr_vol = data[i, j]
        for k, roi_name in enumerate(roi_names):
            mask = rois[roi_name]
            masked_vals = sfnr_vol[mask]
            mean_sfnr[i, j, k] = masked_vals.mean() if masked_vals.size > 0 else np.nan

# mean_sfnr[subject, run, roi]
```

```python
# Create box plot of mean SFNR per ROI
fig, ax = plt.subplots(figsize=(12, 6))

# Prepare data for box plot: reshape mean_sfnr to combine subjects and runs
# mean_sfnr shape: (subjects, runs, rois)
data_for_plot = mean_sfnr.reshape(-1, len(roi_names))

# Create box plot
bp = ax.boxplot([data_for_plot[:, i] for i in range(len(roi_names))], 
                 tick_labels=roi_names,
                 patch_artist=True)

# Customize appearance
for patch in bp['boxes']:
    patch.set_facecolor('lightblue')

ax.set_xlabel('ROI', fontsize=12)
ax.set_ylabel('Mean SFNR', fontsize=12)
ax.set_title('Mean SFNR Distribution per ROI', fontsize=14)
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Create box plot of mean SFNR per run (whole brain)
fig, ax = plt.subplots(figsize=(10, 6))

# Prepare data for box plot: for each run, compute whole brain mean across subjects
data_for_plot = []
for j in range(len(run_labels)):
    # Extract data for this run: all subjects, whole brain
    run_data = [data[i, j][data[i, j] > 0].mean() for i in range(len(subjects))]
    data_for_plot.append(run_data)

# Create box plot
bp = ax.boxplot(data_for_plot, 
                tick_labels=run_labels,
                patch_artist=True)

# Customize appearance
for patch in bp['boxes']:
    patch.set_facecolor('lightblue')

ax.set_xlabel('Run', fontsize=12)
ax.set_ylabel('Mean SFNR', fontsize=12)
ax.set_title('Whole Brain Mean SFNR Distribution by Run', fontsize=14)
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Save mean SFNR volumes as NIfTI images
# Create mean volumes for each run and overall mean
mean_sfnr_run1 = np.mean(data[:, 0, :, :, :], axis=0)
mean_sfnr_run2 = np.mean(data[:, 1, :, :, :], axis=0)
mean_sfnr_run3 = np.mean(data[:, 2, :, :, :], axis=0)
mean_sfnr_overall = np.mean(data, axis=(0, 1))

# Create NIfTI images using the reference image header and affine

mean_run1_img = image.new_img_like(first_img, mean_sfnr_run1)
mean_run2_img = image.new_img_like(first_img, mean_sfnr_run2)
mean_run3_img = image.new_img_like(first_img, mean_sfnr_run3)
mean_overall_img = image.new_img_like(first_img, mean_sfnr_overall)

# Save the images (placeholder paths - you'll add the actual output path)
mean_run1_img.to_filename(os.path.join(outdir, 'mean_sfnr_run1.nii.gz'))
mean_run2_img.to_filename(os.path.join(outdir, 'mean_sfnr_run2.nii.gz'))
mean_run3_img.to_filename(os.path.join(outdir, 'mean_sfnr_run3.nii.gz'))
mean_overall_img.to_filename(os.path.join(outdir, 'mean_sfnr_overall.nii.gz'))
```
